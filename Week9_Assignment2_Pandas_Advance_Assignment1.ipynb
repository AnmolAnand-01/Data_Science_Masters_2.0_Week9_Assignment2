**Q1. List any five functions of the pandas library with execution.**

1. "Series()": It accepts a python list, numpy array, dictionary etc and returns a pandas Series.

```python
import pandas as pd
import numpy as np

#creating a python list
lst=[i for i in range(1,6)]

#creating a numpy array
arr=np.array([9,8,7,6,5])

#creating a dictionary
dct={'a':10,
     'b':20,
     'c':30,
     'd':40,
     'e':50}

#Creating the series using python list
s1=pd.Series(lst)

#Creating the series using numpy array
s2=pd.Series(arr)

#Creating the series using dictionary
s3=pd.Series(dct)

print(s1)
print(s2)
print(s3)
```

2. "DataFrame()": It accepts a python list, numpy array, dictionary etc and returns a pandas DataFrame.

```python
import pandas as pd
import numpy as np

#creating a python list
lst=[i for i in range(1,6)]

#creating a numpy array
arr=np.array([9,8,7,6,5])

#creating a dictionary
dct={'a':[10,1,2,3,4],
     'b':[20,1,2,3,4],
     'c':[30,1,2,3,4],
     'd':[40,1,2,3,4],
     'e':[50,1,2,3,4]}

#Creating the series using python list
s1=pd.DataFrame(lst)

#Creating the series using numpy array
s2=pd.DataFrame(arr)

#Creating the series using dictionary
s3=pd.DataFrame(dct)

print(s1)
print('\n')
print(s2)
print('\n')
print(s3)
```

3. fillna() : Used for handling Missing Values.

```python
import pandas as pd
import numpy as np

# Creating a DataFrame with missing values
data = {'A': [1, 2, np.nan, 4, 5],
        'B': [6, np.nan, 8, np.nan, 10],
        'C': [np.nan, 12, 13, 14, 15]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Fill missing values with a specific value (e.g., 0)
df_filled = df.fillna(0)
print("\nDataFrame after filling missing values:")
print(df_filled)
```

4. dropna(): Drops rows or columns containing missing values based on specified parameters

```python
import pandas as pd
import numpy as np

data = {'A': [1, 2, np.nan, 4, 5],
        'B': [6, np.nan, 8, np.nan, 10],
        'C': [np.nan, 12, 13, 14, 15]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Drop rows with any missing values
df_dropped = df.dropna()
print("\n",df_dropped)

# Drop columns with any missing values
df_dropped_cols = df.dropna(axis=1)
print("\n",df_dropped_cols)
```

5. apply() - Applying Functions.

```python
# Applying a custom function to a column
data = {'Numbers': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Define a custom function to square the values
def square(x):
    return x ** 2

# Applying the 'square' function to the 'Numbers' column and store the result in a new column 'Squared_Numbers'
df['Squared_Numbers'] = df['Numbers'].apply(square)
print("\nDataFrame after applying custom function:")
print(df)
```

**Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.**

```python
import pandas as pd

data={'A':[10,1,2,3,4],
     'B':[20,1,2,3,4],
     'C':[30,1,2,3,4]}

df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)

#Python function to reindex the dataframe.
def reIndexDf(df):
    new_index=[i for i in range(1,len(df)*2,2)]
    new_df=df.reindex(index=new_index)
    return new_df

reindexed_df=reIndexDf(df)
print("DataFrame after reindexing:")
print(reindexed_df)
```

**Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.**

**For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
calculate and print the sum of the first three values, which is 60.**

```python
import pandas as pd
import numpy as np

data={'Values':[10,np.nan,2,np.nan,4],
     'B':[20,1,2,3,4]}
df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)

def CalculateSum(df):
    if len(df)<3:
        print("Values column has less than 3 numbers")
    sum_=0
    cnt=0
    for column_name,series in df.items():
        if column_name=='Values':
            for i in series:
                if cnt<3 and not np.isnan(i):
                    sum_+=i
                    cnt+=1
    if cnt==3:
        print("sum:",sum_)  
    else:
        print("The column 'Values' has less than 3 numerical values")
        

CalculateSum(df)
```

In the above program function "CalculateSum()" calculates the sum of first 3 numerical values present in the column "Values" of the dataframe.

"np.isnan(i)" returns "True" if the value(i) passed to it is "nan(Not A Number)".

**Q5. How are DataFrame.size() and DataFrame.shape() different?**

**1. DataFrame.size:**

DataFrame.size returns the total number of elements (cells) in the DataFrame, calculated by multiplying the number of rows by the number of columns.

It represents the count of all elements within the DataFrame, including both null and non-null values.


**2. DataFrame.shape:**

DataFrame.shape returns a tuple representing the dimensions of the DataFrame.

It provides the number of rows and columns in the DataFrame in the form of 
(number_of_rows, number_of_columns).

```python
import pandas as pd

data={'A':[10,11,12,13,14],
     'B':[20,21,22,23,24],
     'C':[30,31,32,33,34]}

df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("Size: ",df.size)
print("Shape :",df.shape)
# df.size()
```

**Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
function should return a new DataFrame that contains only the selected rows.**

**For example, if df contains the following values:**

**. A B C**

**0 3 5 1**

**1 8 2 7**

**2 6 9 4**

**3 2 3 5**

**4 9 1 2**

**Your function should select the following rows:**

**. A B C**

**1 8 2 7**

**4 9 1 2**

**The function should return a new DataFrame that contains only the selected rows.**

```python
import pandas as pd

data={'A':[3,8,6,2,9],
      'B':[5,2,9,3,1],
      'C':[1,7,4,5,2]}

df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)

def filter_df(df):
    df1=df[(df['A']>5) & (df['B']<10)]
    return df1

new_df=filter_df(df)
print("\nNew DataFrame:")
print(new_df)
```

**Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
median, and standard deviation of the values in the 'Values' column.**

```python
import pandas as pd
import numpy as np

data={'Values':[10,11,11,13,4,8],
     'B':[20,1,2,3,4,8]}
df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)

def mean_median_mode(df):
    return df['Values'].mean(), df['Values'].median(), df['Values'].std()

mean_,median_,mode_=mean_median_mode(df)
print("\nMean: ",mean_)
print("Median: ",median_)
print("Standard Deviation: ",mode_)
```

**Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.**

```python
import pandas as pd

date=pd.date_range("2023-01-11","2023-01-20")
sales=[i*2 for i in range(101,121,2)]
data={"Sales":sales,
      "Date":date}
df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)

def avg_sales(df):
    df['MovingAverage']=df['Sales'].rolling(window=7).mean()
    return df

new_df=avg_sales(df)
print("\nDataFrame after adding the new column:")
print(new_df)
```

**Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.**

**For example, if df contains the following values:**

**. Date**

**0 2023-01-01**

**1 2023-01-02**

**2 2023-01-03**

**3 2023-01-04**

**4 2023-01-05**

**Your function should create the following DataFrame:**

**. Date       Weekday**

**0 2023-01-01 Sunday**

**1 2023-01-02 Monday**

**2 2023-01-03 Tuesday**

**3 2023-01-04 Wednesday**

**4 2023-01-05 Thursday**

**The function should return the modified DataFrame.**

```python
import pandas as pd

date=pd.date_range("2023-01-11","2023-01-20")
data={"Date":date}
df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)

def add_dayName(df):
    df['Weekday']=df['Date'].dt.day_name()
    return df

new_df=add_dayName(df)
print("\nDataFrame after adding the new column:")
print(new_df)
```
